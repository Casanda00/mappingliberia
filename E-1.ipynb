## Getting started!

```python
import ee
import geemap
```

**Great!**

The next step would be to **create an empty map**:
In the map view, you will see search icon where you can use to search for data and so on.

```python
m = geemap.Map()
m
```

**Nice!** 

Now we have started! We can search for the data we want in the search bar. Assuming we searched for elevation data, the source code to import the data is availabe to be copied. See the case below:


```python

dataset = ee.Image('CGIAR/SRTM90_V4')
elevation = dataset.select('elevation')
slope = ee.Terrain.slope(elevation)
m.setCenter(-112.8598, 36.2841, 10)
m.addLayer(slope, {"min": 0, "max": 60}, 'slope')
```

## Another way to source data

We can use the import function from geemap with includes the dot notation. 

```python
# Create the empty map again
m = geemap.Map(center=[40, -100], zoom=4)
m
```

```python
from geemap.datasets import DATA
```

```python
# Using the Dot (.) notation, we can see all the datasets available in geemap. Simply type DATA. "then hit tab"
# This helps us to find the data.
DATA.USGS_GAP_CONUS_2011
```

```python
# Add the layer called GAP CONUS to the empty map.
dataset = ee.Image('USGS/GAP/CONUS/2011')
m.add_layer(dataset, {}, 'GAP CONUS')
```

```python
# Or we can load an entire feature collection of images. Remember to look at the documentation for the type of image file to see if it is a single image or a collection
m = geemap.Map(center=[40, -100], zoom=4)
dataset = ee.ImageCollection('COPERNICUS/S2_CLOUD_PROBABILITY')
m.add_layer(dataset, {}, 'COPERNICUS')
m
```

## Clipping Images or features

Okay, so we have loaded the image for the entire earth. We are not interested in conducting analysis for all countries on earth. Instead, we have our roi. This is where we need to clip the dataset/image to our area.

We can follow the same principle, that is, create an empty map, add the full world map, then add the clipped data layer

```python
DATA.USGS_GAP_CONUS_2011
```

```python
m = geemap.Map(center=[40, -100], zoom=4)
ex = ee.Image('USGS/GAP/CONUS/2011')
m.add_layer(ex, {}, 'CONUS')
m
```

There are two ways to clip an image. We can **clip to a geometry** or we can **clip to a feature collection.**
The recommended way is to clip to a feature collection.

```python
# To access the functions for clipping. use the variable or object name, then dot notation, and then tab to see the list of functions
# Before clipping, make sure to draw the area on the map so that the data can clipped as a new image when the layer is added to the map.
subset = ex.clipToCollection(m.user_rois)

# Then add the layer to the map
m.add_layer(subset, {}, 'Clipped')
```

```python
# Create an empty map
m = geemap.Map(center=[6.4, -9.4], zoom=7.4)
vis_params = {
    'min': 0,
    'max': 6000,
    'bands': ['B4', 'B3', 'B2']
}
m
```

```python
# Search for data
DATA.COPERNICUS_S2_SR_HARMONIZED
```

```python
DATA.USDOS_LSIB_2017
```

```python
data = ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')

# filter the data
collection = (
    ee.ImageCollection('COPERNICUS/S2_SR_HARMONIZED')
    .filterDate("2024-01-01", "2025-01-01")
    .filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 5))
)

image = collection.median()
# Add layer to map
m.add_layer(image, vis_params, 'Sentinel 2')
```

```python
# Filtering Liberia
country = ee.FeatureCollection('USDOS/LSIB/2017')
m.add_layer(country, {}, 'Countries')
```
